# Plotting

`matplotlib` is a desktop plotting package for creating mostly 2d plots. To work with `matplotlib`, simply import the library as shown below:

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib notebook 
#include this line to get zoom-able, resizeable plots
```

```python
simplePlot = np.arange(10)
simplePlot
```

We can create a simple line graph using the `.plot()` method and the NumPy array in the previous cell.

```python
plt.plot(simplePlot)
```

Plots reside within a Figure object; a new figure can be created by doing the following:

```python
fig = plt.figure()
```

A plot cannot be created with a blank figure so subplots must be created inside the figure. To create subplots within a figure, the `.add_subplot()` method is used on the already created figure object.

The `.add_subplot()` method accepts multiple arguments; for our purposes, the 3 most important are rows, columns and position. The rows/columns arguments specifies the shape of the Figure; the position argument refers to where the subplot will be placed. 

No actual plots(graphs) are created using the `.add_subplot()` method; only the containers for the graphs are created at this point.

```python
first = fig.add_subplot(2,2,1) #the Figure is a 2x2 space with this subplot taking up the first space
second = fig.add_subplot(2,2,2)
third = fig.add_subplot(2,1,2) #this plot will span the entire second row
```

To add a graph to a subplot, the `plot` function (part of the matplotlib.pyplot namespace) can be called. In the code below, a line graph will be added to the last subplot in the active figure. That may be in a cell above the cell which calls the function.

```python
plt.plot(simplePlot, 'k--') #k-- creates a black dashed line
```

To plot on a specific subplot in the *active* Figure, refer to the instance methods of the subplot. In this case (below) we want to create a histogram in the second subplot (the subplot was created and stored in the `second` variable) so the method will be called on that subplot object.

In this histogram plot, a few arguments are supplied:
1. The data to plot along the x axis
2. `bins` which refers to the number of intervals to divide the data into
3. `color` which refers to the color of the bars
4. `alpha` which refers to the opacity of the bars (how transparent they are)

```python
second.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
```

Here we populate the subplot referred to by the `first` variable with a scatter plot using the `scatter` instance method. The arguments passed represent the x- and y-axis coordinates.

```python
first.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
```

### Colors, Markers and Line Styles

Matplotlib's main plot function accepts arrays of x and y arguments and optionally a string abbreviation indicating color and line style. In the plot below, the first two arguments, x and y, are based on NumPy arrays and the last argument is a shorthand notation that specifies the line color should be green and the line style should be dashed. To see a full list of line styles and colors, visit: https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = np.arange(40)
y = np.arange(40)*0.2
y = y.cumsum()
plt.plot(x,y, 'g--')
#written explicitly plt.plot(x,y,linestyle='--',color='g')
```

Line plots can have markers that highlight actual data points. The marker can be included in the shorthand argument that specifies line style and color.

```python
plt.plot(np.random.randn(30).cumsum(),'ko--', label="Default")
#written explicitly color='k', linestyle='dashed', marker='o'
```

The drawstyle argument is used to change the interpolation between datapoints (linear is default). Interpolation refers to how the lines between data points are drawn.

```python
plt.plot(np.random.randn(30).cumsum(),'k', drawstyle='steps-post', label='steps-post')
```

To place a legend, the `legend` function, part of the matplotlib.pyplot namespace, must be called. The label arguments in the prior plot functions are used to identify each plot and the `loc` argument of the `legend` function determines the placement of the legend. Other possible values for the `loc` argument include:  
- upper right	
- upper left	
- lower left	
- lower right	
- right	
- center left
- center right	
- lower center
- upper center	
- center

```python
plt.legend(loc='best') #this must be called to create the legend even if no label arguments are passed
```

### Setting the plot title, axis lables, ticks and tick labels

To change the x-axis ticks, use the `.set_xticks()` method; this sets the location of the ticks. To change the text displayed at each tick, use the `.set_xticklabels()` method. To do the same for the y-axis use the `set_yticks()` and `.set_yticklabels()`.

The labels at each axis can be specified by using the `.set_xlabel()` and `.set_ylabel()`. The title of the plot overall can be set using the `.set_title()` method.

```python
newFig = plt.figure()
ax = newFig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

#To change the x-axis ticks, use set_xtics and set_xticklabels
ticks = ax.set_xticks([0,200,500,750,1000])
label = ax.set_xticklabels(['one','two','three','four','five'], rotation=30, fontsize='small')

#set_xlabel names the x-axis and set_title gives the subplot a name
ax.set_xlabel('Stages')
ax.set_title('Super cool graph!')
```

### Saving to a file

The plot(s) in the active figure can be saved to a file using the `.savefig()` function. The filename, with supported filetype included, is required as an argument. Other arguments that can be passed are `dpi`, which refers to the resolution of the figure; `bbox_inches`, which refers to the portion of the figure that will be saved (bounding box); and `facecolor`, which refers to the background outside the subplots

```python
plt.savefig('firstplot.svg')
plt.savefig('firstplot.jpg')
plt.savefig('firstplot.pdf', dpi=400, bbox_inches="tight", facecolor='g')

#To get a list of supported filetypes
plt.gcf().canvas.get_supported_filetypes()
```

### pandas and Plotting
Series and DataFrame each have a plot attribute for making basic plot types. By default `.plot` makes line plots.

```python
obj = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0,100,10))
obj
```

The indices of the Series are used on the x-axis, the values of the Series are used on the y-axis

```python
obj.plot()
```

```python
obj2 = pd.DataFrame(np.random.randn(10,4).cumsum(0), columns=['A','B','C','D'], index=np.arange(0,100,10))
obj2
```

DataFrame's plot method plots each of its columns on a different line on the same subplot, creating the legend automatically

Method arguments for Series and DataFrame can be found on page 271 of the textbook

```python
obj2.plot()
```

**Scatter Plots/Linear Regression**  
Point/scatter plots can be a useful way of examining the relationship between two one-dimensional data series

```python
#https://github.com/statsmodels/statsmodels/blob/master/statsmodels/datasets/macrodata/macrodata.csv

macro = pd.read_csv('macrodata.csv') #read in the csv file
data = macro[['cpi','m1','tbilrate','unemp']] #Get the cpi, m1, tbilrate and unemp columns from the original DataFrame
data.tail()
```

```python
data.plot() #Plotting the columns of the DataFrame as a line plot
```

To create a linear regression plot from this data, we can use either the seaborn or scikit-learn libraries. The process is similar for both libraries.

```python
scatter_data = np.log(data).diff().dropna()
scatter_data.tail()
```

Seaborn is a statistical graphics library that simplifies creating many common visualization types.

Seaborn's regplot method makes a scatter plot and fits a linear regression line. The three arguments supplied in the function below correspond to x, y and the DataFrame the contains the x and y arguments as columns

```python
import seaborn as sb
sb.regplot('m1','unemp', data=scatter_data) 
plt.title('Changes in log %s versus log %s' %('m1','unemp'))
```
