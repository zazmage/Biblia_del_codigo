```python
vertexList = ['0', '1', '2', '3', '4', '5', '6']
edgeList = [(0,1), (0,2), (1,0) , (1,3) , (2,0) , (2,4) , (2,5) , (3,1), (4,2) , (4,6), (5,2), (6,4)]

def dfs(vertexList, edgeList, start):
    visitedVertex = []
    stack = [start]
    # Una lista con 6 listas vac√≠as al interior
    adjacencyList = [[] for vertex in vertexList]
    #print(edgeList)
    for edge in edgeList:
        adjacencyList[edge[0]].append(edge[1])
    print(adjacencyList)

    while stack:
        print("Antes Stack: ", stack)
        current = stack.pop()
        print("Despues Stack: ", stack)
        #print("Current: ", current)
        for neighbor in adjacencyList[current]:
            #print("Adj list: ",adjacencyList[current])
            #print("Neighbor: ",neighbor)
            if not neighbor in visitedVertex:
                stack.append(neighbor)
                
        visitedVertex.append(current)
    return visitedVertex

dfs(vertexList, edgeList, 0)
#print(dfs(vertexList, edgeList, 0))
```
