```python
import networkx as nx

nodos = ['0', '1', '2', '3', '4', '5', '6']
aristas = [(0,1), (0,2), (1,0) , (1,3) , (2,0) , (2,4) , (2,5) , (3,1), (4,2) , (4,6), (5,2), (6,4)]

G = nx.Graph()

for i in range(7):
    G.add_node(i)

for i in aristas:
    G.add_edge(i[0],i[1])

nx.draw(G, with_labels=1)
```

```python
vertexList = ['0', '1', '2', '3', '4', '5', '6']
edgeList = [(0,1), (0,2), (1,0) , (1,3) , (2,0) , (2,4) , (2,5) , (3,1), (4,2) , (4,6), (5,2), (6,4)]

def dfs(vertexList, edgeList, start):
    visitedVertex = []
    stack = [start]
    # Una lista con 6 listas vacías al interior
    adjacencyList = [[] for i in vertexList]
    # print(adjacencyList)
    for i in edgeList:
        adjacencyList[i[0]].append(i[1])
    print(adjacencyList)

    # Este while funciona mientras el stack no se quede vacío
    while stack:
        print("Antes Stack: ", stack)
        # Este stack funciona como una pila, y la funcion pop retorna el elemento más arriba
        # de la pila, como cuando se lavan platos.
        current = stack.pop()
        print("Despues Stack: ", stack)
        print("Current: ", current)
        # Recorre los nodos adyacentes al nodo que se esta evaluando, en este caso el nodo
        # de la variable "current".
        for vecino in adjacencyList[current]:
            # Este if evalua si los nodos adyacentes al nodo evaluado no se encuentran dentro
            # de la lista "visitedVertex" (osea que no se han visitado aún), 
            # si no se encuentran, se almacenan en la pila
            if not vecino in visitedVertex:
                stack.append(vecino)
        
        # Añade el nodo evaluado a la lista de nodos visitados
        visitedVertex.append(current)
        #print(visitedVertex)

    return visitedVertex

dfs(vertexList, edgeList, 0)
#print(dfs(vertexList, edgeList, 0))
```
