### 1. Hot Dog Stand  
Create a DataFrame that contains the beginning inventory for a hot dog stand: 20 hot dogs, 20 cans of coke, 20 bottles of water and 10 soft pretzels. Each time a sale is made, the user will enter the quantity of items sold and the DataFrame will be updated to reflect the new available inventory. Output the updated DataFrame after each update. The program must allow the user to continue to enter sales until all the items are sold. Once an item's inventory has been depleted, print a message notifying the user that the item can no longer be sold. If the user tries to record the sale of an item after its inventory has been depleted, print a message alerting the user to their mistake. Once all the items have been depleted, print "All items have been sold" and exit the program.

```python
#Answer 1

#Create the DataFrame using 2 lists
items = ['hot dogs', 'coke', 'water','pretzels']
quantity=[20,20,20,10]
inventory = pd.DataFrame(quantity, index=items, columns=['quantity'])

print(inventory)
#Start a loop based on if the total of the values in the DF are greater than 0
while inventory.values.sum() > 0:
    print('Congrats you made a sale!\nPlease record how many of each item was sold.\n')

    if inventory.loc['hot dogs','quantity'] > 0: #If hot dogs haven't been sold out
        hotdog = int(input('HOT DOGS: ')) #Capture the hot dog sale number
        while hotdog > inventory.loc['hot dogs','quantity']: #If the sale number provided is greater than the available units prompt the user for an amount less than or equal to the number of available units
            print('You can only sell', inventory.loc['hot dogs','quantity'],'hot dogs. Try again')
            hotdog = int(input('Hot dogs:\n'))
        inventory.loc['hot dogs','quantity'] = inventory.loc['hot dogs','quantity']-hotdog #Once an appropriate sale number is entered, subtract that number from the amount in the DF
    else:
        print('You have no more hot dogs to sell') #If hot dogs have been sold out, show a corresponding message
    #Repeat the same process for the other items
    if inventory.loc['coke','quantity'] > 0:
        coke = int(input('COKE: '))
        while coke > inventory.loc['coke','quantity']:
            print('You can only sell', inventory.loc['coke','quantity'],'cans of coke. Try again')
            coke = int(input('Cans of coke:\n'))
        inventory.loc['coke','quantity'] = inventory.loc['coke','quantity']-coke
    else:
        print('You have no more cans of coke to sell')
    if inventory.loc['water','quantity'] > 0:
        water = int(input('WATER: '))
        while water > inventory.loc['water','quantity']:
            print('You can only sell', inventory.loc['water','quantity'],'bottles of water. Try again')
            water = int(input('Bottles of water:\n'))
        inventory.loc['water','quantity'] = inventory.loc['water','quantity']-water
    else:
        print('You have no more bottles of water to sell')
    if inventory.loc['pretzels','quantity'] > 0:
        pretzel = int(input('PRETZELS: '))
        while pretzel > inventory.loc['pretzels','quantity']:
            print('You can only sell', inventory.loc['pretzels','quantity'],'soft pretzels. Try again')
            pretzel = int(input('Soft pretzels:\n'))
        inventory.loc['pretzels','quantity'] = inventory.loc['pretzels','quantity']-pretzel
    else:
        print('You have no more soft pretzels to sell')
    print(inventory,"\n") #Print the DF after each sale
print("All items have been sold") #Once all items have sold out print a message
```

```python
#Answer 2 - Based on answer 1, but using iloc and a for loop to condense the process

#Create the DataFrame using a dictionary
itemsDict = {'hot dogs':20,'coke':20,'water':20,'pretzels':10}
inventory = pd.DataFrame(itemsDict.values(), index=itemsDict.keys(), columns=['quantity'])

print(inventory)
while inventory.values.sum() > 0:
    print('Congrats you made a sale!\nPlease record how many of each item was sold.\n')
    for i in np.arange(0,len(inventory)): #Create a loop that will iterate based on the number of rows in the DF
        #Remember that iloc refers to the integer index location of data in the DF and that i will be a different number each time the loop iterates
        #From this point, the logic is the same as in Answer 1
        if inventory.iloc[i,0] > 0: 
            saleqty = int(input(inventory.index[i].upper()+':')) 
            while saleqty > inventory.iloc[i,0]: 
                print('You can only sell',inventory.iloc[i,0],'of this item. Try again')
                saleqty = int(input(inventory.index[i].upper()+':\n'))
            inventory.iloc[i,0] = inventory.iloc[i,0]-saleqty 
        else:
            print('You have no more',inventory.index[i],'to sell')
    print(inventory)
print("All items have been sold")
```

### 2. Word Score Calculator
Create a Series that contains letter and corresponding point values according to the following:
- 1 point: E, A, I, O, N, R, T, L, S, U
- 2 points: D, G
- 3 points: B, C, M, P
- 4 points: F, H, V, W, Y
- 5 points: K
- 8 points: J, X
- 10 points: Q, Z

Using the Series, create a program that allows the user to enter a word and have the total point value of the word returned. The program must allow the user to continue to enter words until they decide to exit the program.


```python
letterPoints = {'E':1,'A':1,'I':1,'O':1,'N':1,'R':1,'T':1,'L':1,'S':1,'U':1, 
                'D':2,'G':2,
                'B':3,'C':3,'M':3,'P':3,
                'F':4,'H':4,'V':4,'W':4,'Y':4,
                'K':5,
                'J':8,'X':8,
                'Q':10,'Z':10}
lpSeries = pd.Series(letterPoints)
more = 'y'
while more == 'Y' or more == 'y':
    word = input('Enter Your Word: ')
    score = 0
    for i in word.upper():
        print(i,'-',lpSeries[i])
        score += lpSeries[i]
    print(word,'-',score,'points')
    more = input('Calculate another word? (Y/N) ')
```

### 3. Daily Step Stats
Create a program that allows a user to enter their daily step count, 7 days at a time (exactly 7, do not give the option to enter less). For the first 7 days of user input, create a NumPy array of steps and output the array and the following:
1. The total number of steps for all 7 days
2. The average number of steps taken
3. The number of miles walked eached day (1 mile = 2000 steps)
4. The total number of miles for all 7 days
5. The highest step count during the 7 days

Allow the user to enter additional weeks, 7 days at a time. Reshape the NumPy array for each 7 day period added; i.e for the first 7 days the array should be 1 dimensional, for the second, the array shape of the array should be (2,7), etc. After data for each new week is entered, output the updated array and the following:
1. The total number of steps for x days, where x is the size of the array
2. The average number of steps taken, both for each week and overall
3. The number of miles walked eached day
4. The total number of miles for each week
5. The total number of miles for all days
6. The highest step count for each week
7. The highest step count overall

```python
def recordSteps():
    tempList = []
    print('Enter your steps for the week')
    for i in range(7):
        tempList.append(int(input('Day '+str(i+1)+': ')))
    return tempList

weeklySteps = recordSteps()
npSteps = np.array(weeklySteps)
print('---------------------------------------------')
print('Total Steps Taken (',npSteps.size,' days): ', npSteps.sum(),'\n', sep='')
print('Average Steps Taken:', npSteps.mean(),'\n')
print('Number of Miles Walked:')
for i in np.arange(0,npSteps.size):
    print('\tDay', i+1,'-',npSteps[i]/2000)
print('\nTotal Miles (',npSteps.size,' days): ', npSteps.sum()/2000,'\n', sep='')
print('Most Steps Taken:',npSteps.max())
print('---------------------------------------------')
more = input('\nAdd another week? (Y/N)')
while more == 'y' or more=='Y':
    print('---------------------------------------------')
    weeklySteps = weeklySteps + recordSteps()
    npSteps = np.array(weeklySteps).reshape((int(len(weeklySteps)/7),7))
    print('---------------------------------------------')
    print('Total Steps Taken (',npSteps.size,' days): ', npSteps.sum(), sep='')
    print('Average Steps Taken, Weekly')
    for i in range(len(npSteps)):
        print('\tWeek',i+1,'-',npSteps[i].mean())
    print('Average Steps Taken, Overall: ',npSteps.mean())
    print('Total Miles Walked, Weekly')
    for i in range(len(npSteps)):
        print('\tWeek',i+1,'-',npSteps[i].sum()/2000)
    print('Total Miles Walked, Overall: ',npSteps.sum()/2000)
    print('Most Steps Taken, Weekly')
    for i in range(len(npSteps)):
        print('\tWeek',i+1,'-',npSteps[i].max())
    print('Most Steps Taken, Overall:', npSteps.max())
    print('---------------------------------------------')
    more = input('\nAdd another week? (Y/N)')
print('Keep stepping!')
```

### 4. World Happiness
Using the data from the [2015 World Happiness Report](2015.csv), create one figure containing 3 subplots. The first should plot the Happiness Scores for the top 10 countries. The second should plot the Happiness Scores of the bottom 10 countries. Top 10 and bottom 10 should be determined by Happiness Rank, not necessarily the way the data loads from the file. The third should plot the average Happiness Score by Region. Each plot should have an appropriate title, x-/y-axis labels and ticks/tick labels for each axis. Below the figure, output the data (as DataFrames) used to create each of the 3 subplots.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

happiness = pd.read_csv('2015.csv')

topHappy = happiness.drop(happiness.index[happiness['Happiness Rank']>10]) #Remove rows from the original DataFrame if Happiness Rank is greater than 10
bottomHappy = happiness.drop(happiness.index[happiness['Happiness Rank'] < happiness['Happiness Rank'].max()-9]) #Remove rows from the original DataFrame if Happiness Rank is less than 10 less than the maximum Happiness Rank

#To get the mean score by region
regions = list(set(happiness['Region'])) #Convert the Region column to a set to get unique values then convert that set into a list to interate

#Create a dictionary where the keys are the Regions and the values are the mean Happiness Scores for that region
happinessMean = {}
for i in range(len(regions)):
    happinessMean[regions[i]] = happiness.loc[happiness['Region'] == regions[i],'Happiness Score'].mean()
#Use the dictonary to create a new DataFrame
meanDF = pd.DataFrame(happinessMean.values(), index=regions, columns=['Average Happiness Score'])

#Create the figures/subplots
figure = plt.figure()
figure.tight_layout()
top10 = figure.add_subplot(2,2,1)
bot10 = figure.add_subplot(2,2,2)
mean = figure.add_subplot(3,1,3)

#Create the plots and set title, x-/y-labels, x-tick labels
top10.plot(topHappy['Country'],topHappy['Happiness Score'], drawstyle='steps-post')
top10.set_title('2015 Happiness Rank - Top 10')
top10.set_xlabel('Country')
top10.set_xticklabels(topHappy['Country'],rotation=90,fontsize='small')
top10.set_ylabel('Score')

bot10.plot(bottomHappy['Country'],bottomHappy['Happiness Score'])
bot10.set_title('2015 Happiness Rank - Bottom 10')
bot10.set_xlabel('Country')
bot10.set_xticklabels(bottomHappy['Country'],rotation=90,fontsize='small')
bot10.set_ylabel('Score')

mean.bar(meanDF.index,height=meanDF['Average Happiness Score'])
mean.set_title('2015 Average Happiness Score')
mean.set_xlabel('Region')
mean.set_xticklabels(meanDF.index,rotation=90,fontsize='small')
mean.set_ylabel('Avg Score')
```
