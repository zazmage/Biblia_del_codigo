**Filtrando datos**

Filtrar filas según un parámetro

```python
df.loc[df['Type 1'] == "Grass"] # Muestra solo las filas que poseen pokemones tipo grass
```

Filtrar según varios parámetros

```python
# Muestra las filas que contienen pokemones tipo grass y poison
# En pandas cuando realizamos filtrados de más de un parámetro
# Se utiliza '&' en lugar de 'and' y '|' en lugar de 'or'
df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == 'Poison')]
```

Filtrar según varios parámetros con condición de mayor o menor que

```python
# Filtra según los parámetros anteriores añadiendo los pokemones con vida mayor que 70
df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
```

Guardar un nuevo dataframe con el índice reseteado

```python
# Crea un nuevo dataframe a partir del filtrado anterior llamado new_df
new_df = df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

# - Hace un reset a los valores del índice del nuevo dataframe
# - drop = True elimina la opción por defecto de mantener una columna
# con los valores antiguos del index
# - inplace = True nos evita tener que sobreescribir la variable new_df
# de tal forma que "new_df = new_df.reset_index(drop=True)" es innecesario
# puesto que los cambios se hacen directamente sobre la variable
new_df.reset_index(drop=True, inplace = True)
new_df
```

Filtrar por strings que contengan un texto determinado

```python
# Filtra todas las filas cuyo dato nombre contenga la palabra 'Mega'
df.loc[df['Name'].str.contains('Mega')]
```

Filtrar por strings que **NO** contengan un texto determinado

```python
# Filtra todas las filas cuyo dato nombre contenga la palabra 'Mega'
# Pandas en lugar de 'not' y '!' utiliza '~' (leguleya) para negar una sentencia
df.loc[~df['Name'].str.contains('Mega')]
```

Filtrar usando expresiones regulares

```python
import re

# -Filtra los pokemones que sean de tipo fire y grass
# -flags = re.I significa "ignore case" nos permite escribir los parámetros sin preocuparnos
# por las mayusculas o minusculas
# regex = True activa las expresiones regulares para el filtrado
df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

Filtrar datos con iniciales determinadas usando expresiones regulares

```python
# Para mayor información leer documentación sobre expresiones regulares
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

**Cambios condicionales**

Cambiar todas las apariciones de un dato determinado en una columna por otro

```python
# Cambia todos los datos 'Fire' de la columna Type 1 por 'Flamer'
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
# Regresa a los parámetros iniciales
df.loc[df['Type 1'] == 'Flamer', 'Type 1'] = 'Fire'
df
```

Cambiar todas las apariciones de un dato determinado en una columna según una condición en otra

```python
# Todo pokemon de tipo Fire sera de tipo Legendary
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True
df
```

**Estadística agregada (agrupar por)**

Promediar todos los datos de las filas donde aparezca un dato en común

```python
# Promedia todos los datos de los pokemones según su tipo
df.groupby(['Type 1']).mean()
```

Promediar todos los datos de las filas donde aparezca un dato en común de forma ascendente o descendente

```python
# Promedia todos los datos de los pokemones según su tipo y los organiza de forma descendente
# según el valor de su 'Attack'
df.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

Sumar todos los datos de las filas donde aparezca un dato en común

```python
# Suma todos los datos de los pokemones según su tipo
df.groupby(['Type 1']).sum()
```

Contar todos los datos de las filas donde aparezca un dato en común

```python
# Cuenta todos los datos de los pokemones según su tipo
df.groupby(['Type 1']).count()
```

Contar todos los datos de las filas donde aparezca un dato en común y organizarlos según varios parámetros

```python
# Crea una nueva columna count que valdra 1 en todos sus datos
df['Count'] = 1

# Muestra los datos organizados en todos los valores de 'Type 2' que hay para cada 'Type 1'
df.groupby(['Type 1', 'Type 2']).count()['Count']
```

**Trabajando en grandes volumenes de datos**

Cargar una parte parcial o reducida de un archivo con un gran volumen de datos

```python
# - Carga el dataset previamente creado pero simulando que es muy grande como para que nuestra
# computadora pueda cargarlo completo, por lo que vamos a cargar una versión parcializada
# - Por cada iteración del ciclo for carga una parte del 'chunk' o 'pedazo' del dataframe y
# ejecuta el código correspondiente a cada elemento del dataframe
# - chunksize = 5 determina cuantas filas se van a cargar en memoria por cada iteración
for df in pd.read_csv('modified_sin_indices.csv', chunksize = 5):
    print("CHUNK OF") # Imprime 'CHUNK OF' cada que carga una parte nueva del dataset
    print(df)
```

Trabajando con dataframes fragmentados

```python
# Crea un nuevo dataframe que solo almacena los indices del dataframe original de pokemones
new_df = pd.DataFrame(columns=df.columns)
# Recorre el dataframe por secciones en cada iteración
for df in pd.read_csv('modified_sin_indices.csv', chunksize = 5):
    # Almacena en resoults el valor del conteo de datos de cada columna de cada fragmento de dataframe
    resoults = df.groupby(['Type 1']).count()
    # Añade al dataframe el resultado de cada iteración
    new_df = pd.concat([new_df, resoults])
new_df
```
